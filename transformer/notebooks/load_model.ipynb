```python
import os
import wandb

save_path = "/scratch/htc/npelleriti/models/sos-transformer/"

# configure your project
entity = "ais2t"   
project = "sos-transformer"  
sweep_id = "j4pe9eny"

api = wandb.Api()

# download all artifacts from the sweep
sweep = api.sweep(f"{entity}/{project}/{sweep_id}")
runs = sweep.runs
for run in runs:
    artifacts = run.logged_artifacts()
    for art in artifacts:
        # Get the data_path from run config to create subdir
        data_path = run.config.get('data_path', 'default')
        # Extract only the last part of the data_path
        data_subdir = os.path.basename(data_path)
        target_dir = os.path.join(save_path, data_subdir)
        print(target_dir)
        os.makedirs(target_dir, exist_ok=True)
        art.download(root=target_dir)
```
